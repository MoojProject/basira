```python
def keep_row(field_name, keep_index):
    global df

    rows_before = len(df)

    df = df.drop(
        df[(df["field"] == field_name) & (df.index != keep_index)].index
    )

    rows_after = len(df)

    print(f"تم تنظيف {field_name}")
    print("تم حذف:", rows_before - rows_after, "صف")
def review_field(field_name):
    rows = df[df["field"] == field_name].copy()

    if len(rows) == 0:
        return

    print("=" * 100)
    print("FIELD:", field_name)
    print("عدد الصفوف:", len(rows))
    print("=" * 100)

    cols = [
        "field",
        "title_ar",
        "category",
        "required",
        "risk_if_missing",
        "description",
        "keywords",
        "reference"
    ]

    print(rows[cols].to_string())
```
